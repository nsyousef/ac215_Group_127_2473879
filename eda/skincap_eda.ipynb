```python
# Image sources
print(df['source'].value_counts())
df['source'].value_counts().plot.bar()
```

```python
# missing data
print(df.isnull().sum().to_string())
```

```python
# Label distribution
import matplotlib.pyplot as plt

df_disease_count = df['disease'].value_counts()
print(f'{len(df_disease_count)} diseases present')

plt.figure(figsize=(25,6))
df_disease_count.plot.bar()
plt.tight_layout()
plt.show()
```

```python
# Low representation diseases
df_low_representation = df_disease_count[df_disease_count < 5]
print(f'{len(df_low_representation)} diseases have less than 5 images')
print(df_low_representation)
```

```python
# Missing data from DDI vs Fitzpatrick17k images
print(f'Fitzpatrick missing data:\n{df[df['source'] == 'fitzpatrick17k'].isnull().sum().to_string()}')
print(f'\nDDI missing data:\n{df[df['source'] == 'ddi'].isnull().sum().to_string()}')
```

```python
# Metadata exclusive to DDI vs. Fitzpatrick17k
print('DDI:')
print(df['skin_tone'].value_counts())
print(df['malignant'].value_counts())

print('\nFitzpatrick 17k')
print(df['fitzpatrick_scale'].value_counts())
print(df['fitzpatrick_centaur'].value_counts())
print(df['nine_partition_label'].value_counts())
print(df['three_partition_label'].value_counts())
```
